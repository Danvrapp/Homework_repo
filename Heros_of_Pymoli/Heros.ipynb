```python
#Import libraries
import pandas as pd
import numpy as np
import csv
import os
```

```python
#Read in file
Hero_file = '/Users/Dan/Desktop/GIT_Homework/Homework/Heros_of_Pymoli/Heros.csv'

#Determine number of unique users
Heros = pd.read_csv(Hero_file)

#rename column header(s)
Heros = Heros.rename(columns={"SN":"Screen_name"})
Users = len(Heros["Screen_name"].unique())
print("There are", Users, "unique users")
```

```python
List_of_users = Heros.iloc[:, 1]
Unique_users_df = pd.DataFrame({"Screen_name":List_of_users.unique()})
Unique_users_df.head()
```

```python
#Calculate and display average purchase price for the df
Items = len(Heros["Item ID"].unique())
average_price = "$" + "%0.2f" %Heros["Price"].mean()
num_of_buys = len(Heros["Purchase ID"].value_counts())
revenue = "$" + "%0.2f" %Heros["Price"].sum()
purchasing_df = pd.DataFrame({"Unique Items":[Items],
                              "Average Price":[average_price],
                              "Number of purchases":[num_of_buys],
                              "Total revenue":[revenue]})
purchasing_df
```

```python
#calculate gender of players
male_df = Heros.loc[Heros["Gender"]=="Male", :]
female_df = Heros.loc[Heros["Gender"]=="Female", :]
other_df = Heros.loc[Heros["Gender"]=="Other / Non-Disclosed", :]
unique_males = len(male_df["Screen_name"].unique())
unique_other = len(other_df["Screen_name"].unique())
unique_females = len(female_df["Screen_name"].unique())

gender_df = pd.DataFrame({"Gender category":["Male", "Female", "Other / Non-disclosed"],
                         "Number of":[unique_males, unique_females, unique_other],
                         "Percentage":["%0.2f" %(unique_males * 100/Users), "%0.2f" %(unique_females*100/Users), "%0.2f" %(unique_other*100/Users)]})
gender_df
```

```python
females_df = Heros.loc[Heros["Gender"]=="Female"]
males_df = Heros.loc[Heros["Gender"]=="Male"]
other_df = Heros.loc[Heros["Gender"]=="Other / Non-Disclosed"]
```

```python
females_grouped_df = females_df.groupby(["Screen_name"])
males_grouped_df = males_df.groupby(["Screen_name"])
other_grouped_df = other_df.groupby(["Screen_name"])
```

```python
fem_pur = len(females_df["Purchase ID"])
fem_avg_pur_price = females_df["Price"].mean()
fem_tot_pur = females_df["Price"].sum()
fem_avg_person_pur = fem_tot_pur / len(females_grouped_df)
mal_pur = len(males_df["Purchase ID"])
mal_avg_pur_price = males_df["Price"].mean()
mal_tot_pur = males_df["Price"].sum()
mal_avg_person_pur = mal_tot_pur / len(males_grouped_df)
oth_pur = len(other_df["Purchase ID"])
oth_avg_pur_price = other_df["Price"].mean()
oth_tot_pur = other_df["Price"].sum()
oth_avg_person_pur = oth_tot_pur / len(other_grouped_df)
purchase_df = pd.DataFrame({"Gender":["Female", "Male", "Other / Non-Disclosed"],
                           "Purchase Count":[fem_pur, mal_pur, oth_pur],
                           "Average Purchase Price":["%0.2f" %fem_avg_pur_price,"%0.2f" %mal_avg_pur_price,"%0.2f" %oth_avg_pur_price],
                           "Total Purchase Value":["%0.2f" %fem_tot_pur,"%0.2f" % mal_tot_pur,"%0.2f" % oth_tot_pur],
                           "Average Total Purchase per Person":["%0.2f" %fem_avg_person_pur,"%0.2f" % mal_avg_person_pur,"%0.2f" % oth_avg_person_pur]})
purchase_df.head()
```

```python
age_group_bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 100]
age_group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "34-39", "40+"]
Heros["Age Groups"] = pd.cut(Heros["Age"], age_group_bins, labels=age_group_names)
```

```python
Under_10 = len(Heros.loc[Heros["Age"]<10, ["Screen_name", "Age"]]["Screen_name"].unique())
Under_15 = len(Heros.loc[(Heros["Age"]<15) & (Heros["Age"]>9), ["Screen_name", "Age"]]["Screen_name"].unique())
Under_20 = len(Heros.loc[(Heros["Age"]<20) & (Heros["Age"]>14), ["Screen_name", "Age"]]["Screen_name"].unique())
Under_25 = len(Heros.loc[(Heros["Age"]<25) & (Heros["Age"]>19), ["Screen_name", "Age"]]["Screen_name"].unique())
Under_30 = len(Heros.loc[(Heros["Age"]<30) & (Heros["Age"]>24), ["Screen_name", "Age"]]["Screen_name"].unique())
Under_35 = len(Heros.loc[(Heros["Age"]<35) & (Heros["Age"]>29), ["Screen_name", "Age"]]["Screen_name"].unique())
Under_40 = len(Heros.loc[(Heros["Age"]<40) & (Heros["Age"]>34), ["Screen_name", "Age"]]["Screen_name"].unique())
Over_40 = len(Heros.loc[Heros["Age"]>39, ["Screen_name", "Age"]]["Screen_name"].unique())
Total = Under_10 + Under_15 + Under_20 + Under_25 + Under_30 + Under_35 + Under_40 + Over_40
Age_df = pd.DataFrame({"Age group": age_group_names, 
                      "Total_count": [Under_10, Under_15, Under_20, Under_25, Under_30, Under_35, Under_40, Over_40],
                      "Percentage of Players": ["%0.2f" %(Under_10*100/Total), "%0.2f" %(Under_15*100/Total), "%0.2f" %(Under_20*100/Total), "%0.2f" %(Under_25*100/Total), "%0.2f" %(Under_30*100/Total), "%0.2f" %(Under_35*100/Total), "%0.2f" %(Under_40*100/Total), "%0.2f" %(Over_40*100/Total)]})
Age_df.head(10)
```

```python
Under_10_sales = Heros.loc[Heros["Age"]<10, ["Price"]].count()
Under_15_sales = Heros.loc[(Heros["Age"]<15) & (Heros["Age"]>9), ["Price"]].count()
Under_20_sales = Heros.loc[(Heros["Age"]<20) & (Heros["Age"]>14), ["Price"]].count()
Under_25_sales = Heros.loc[(Heros["Age"]<25) & (Heros["Age"]>19), ["Price"]].count()
Under_30_sales = Heros.loc[(Heros["Age"]<30) & (Heros["Age"]>24), ["Price"]].count()
Under_35_sales = Heros.loc[(Heros["Age"]<35) & (Heros["Age"]>29), ["Price"]].count()
Under_40_sales = Heros.loc[(Heros["Age"]<40) & (Heros["Age"]>34), ["Price"]].count()
Over_40_sales = Heros.loc[Heros["Age"]>39, ["Price"]].count()

U10_pur_df = Heros.loc[Heros["Age"]<10, ["Price"]].sum()
U15_pur_df = Heros.loc[(Heros["Age"]>9)&(Heros["Age"]<15), ["Price"]].sum()
U20_pur_df = Heros.loc[(Heros["Age"]>14)&(Heros["Age"]<20), ["Price"]].sum()
U25_pur_df = Heros.loc[(Heros["Age"]>19)&(Heros["Age"]<25), ["Price"]].sum()
U30_pur_df = Heros.loc[(Heros["Age"]>24)&(Heros["Age"]<30), ["Price"]].sum()
U35_pur_df = Heros.loc[(Heros["Age"]>29)&(Heros["Age"]<35), ["Price"]].sum()
U40_pur_df = Heros.loc[(Heros["Age"]>34)&(Heros["Age"]<40), ["Price"]].sum()
O40_pur_df = Heros.loc[Heros["Age"]>39, ["Price"]].sum()
Age_pur_df = pd.DataFrame({"Age group": age_group_names,
                           "Total Purchase Value": ["$" + "%0.2f" %(U10_pur_df), "$" + "%0.2f" %(U15_pur_df), "$" + "%0.2f" %(U20_pur_df), "$" + "%0.2f" %(U25_pur_df), "$" + "%0.2f" %(U30_pur_df), "$" + "%0.2f" %(U35_pur_df), "$" + "%0.2f" %(U40_pur_df), "$" + "%0.2f" %(O40_pur_df)],
                           "Purchase Count":["%0.0f" %Under_10_sales, "%0.0f" %Under_15_sales, "%0.0f" %Under_20_sales, "%0.0f" %Under_25_sales, "%0.0f" %Under_30_sales, "%0.0f" %Under_35_sales, "%0.0f" %Under_40_sales, "%0.0f" %Over_40_sales ],
                           "Average Purchase Price":["$" + "%0.2f" %(U10_pur_df/Under_10_sales), "$" + "%0.2f" %(U15_pur_df/Under_15_sales), "$" + "%0.2f" %(U20_pur_df/Under_20_sales), "$" + "%0.2f" %(U25_pur_df/Under_25_sales), "$" + "%0.2f" %(U30_pur_df/Under_30_sales), "$" + "%0.2f" %(U35_pur_df/Under_35_sales), "$" + "%0.2f" %(U40_pur_df/Under_40_sales), "$" + "%0.2f" %(O40_pur_df/Over_40_sales)],
                           "Avg. Purchase per person":["$" + "%0.2f" %(U10_pur_df/Under_10), "$" + "%0.2f" %(U15_pur_df/Under_15), "$" + "%0.2f" %(U20_pur_df/Under_20), "$" + "%0.2f" %(U25_pur_df/Under_25), "$" + "%0.2f" %(U30_pur_df/Under_30), "$" + "%0.2f" %(U35_pur_df/Under_35), "$" + "%0.2f" %(U40_pur_df/Under_40), "$" + "%0.2f" %(O40_pur_df/Over_40)]})
Age_pur_df.head(10)
```
